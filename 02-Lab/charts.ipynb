```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
# Box Graph
sns.boxplot(x="Geography", y="Balance", data=dataset,hue="Gender")
```

```python
# Scatter plot
sns.scatterplot(x="Age", y="Balance", data=dataset, hue="Gender")
```

```python
# Line Graphs
sns.lineplot(x="Age", y="Balance", data=dataset, hue="Gender")
```

```python
# Bar Chart
sns.barplot(x="Geography",y="Balance",data=dataset,hue="Gender")
```

```python
# Histograms
sns.histplot(data=dataset, x="Geography",hue="Gender")
```

```python
data = dataset
```

```python
data = dataset.groupby(['Geography']).sum()
```

```python
data
```

```python
balance = data["HasCrCard"]
```

```python
balance
```

```python
creditCard = []

sum = 0
for i in balance: 
    creditCard.append(i)
    sum = sum + i

creditCard.append(10000-sum)
```

```python
creditCard
```

```python
labels = dataset['Geography'].unique()
```

```python
label = []
for i in labels:
    label.append(i)

label.append("Not Holding any credit card")
```
