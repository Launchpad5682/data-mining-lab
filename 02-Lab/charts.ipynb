```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
# Box Graph
sns.boxplot(x="Geography", y="Balance", data=dataset,hue="Gender")
```

```python
# Scatter plot
sns.scatterplot(x="Age", y="Balance", data=dataset, hue="Gender")
```

```python
# Line Graphs
sns.lineplot(x="Age", y="Balance", data=dataset, hue="Gender")
```

```python
# Bar Chart
sns.barplot(x="Geography",y="Balance",data=dataset,hue="Gender")
```

```python
# Histograms
sns.histplot(data=dataset, x="Geography",hue="Gender")
```
